```python
# KERAS BASICS
# 1. Sequential API  - Beginner level
# 2. Functional API  - Intermediate level
#  Sequential API - model is created layer by layer
#  Functional APi - model is created all at once
```

```python
# Activation Layer: which specifies the activation function
# Dense Layer: every neuron is connected to every neuron of its immediate next layer
# Conv2D : convolutional layer
# MaxPooling layer: Downsampling(reducing number of features)
```

```python
# MODEL BUILDING in KERAS

# 1. Gather the data (df, preprocessing, handling missing data, eda, etc.)
# 2. using Sequential API, create NN architecture (create Artificial Neurons)
#  -Input Layer : if the input layer is in vector form, we do not need to flatten it
#  -Hidden Layer : input_shape, activation function
#  -Output Layer : activation function
# 3. Compile : loss function, optimizers(SGD, Adam,ec.),  metrics(accuracy, r2_score,etc.)
# 4. Fit the model : epochs, batch_size, etc.
# 5. Predict the new values and evaluate the model
```

```python
#  Rules regarding the activation function
# For input layer, no activation function
# For op layer, the activation function will depend
        # For regression    use relu or linear
        # For Classification
                # Binary                    Sigmoid
                # Multi class classifer     Softmax
```

```python
# LOSS function for classification
# BCE (Binary Cross Entropy)
```

```python
from math import log

p=[1,1,1,1,0,0,0,0]  # actual op
q=[0.8,0.75,0.7,0.5,0.3,0.3,0.4,0.1]  # pred op

def binary_ce(p,q):
    return -((p*log(q)))+((1-p)*log(1-q))
```

```python
result=[]
for i in range(len(p)):
    ce=binary_ce(p[i],q[i])
    result.append(ce)

mean_ce=np.mean(result)

mean_ce
```

```python
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense,Dropout
from keras.optimizers import RMSprop
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

```python
x_train=x_train.reshape(len(x_train),28*28)
x_test=x_test.reshape(len(x_test),28*28)
```

```python
x_train=x_train.astype('float32')
x_test=x_test.astype('float32')
```

```python
x_train /= 255   #x_train=x_train/255
x_test /= 255
```

```python
num_classes=10
y_train=keras.utils.to_categorical(y_train,num_classes)
y_test=keras.utils.to_categorical(y_test,num_classes)
```

```python
y_train[333]
```

```python
y_train.shape
```

```python
y_train[333].shape
```

```python
model1=Sequential()
model1.add(Dense(64,activation='relu',input_shape=(784,)))
model1.add(Dropout(0.2))
model1.add(Dense(64,activation='relu'))
model1.add(Dropout(0.2))
model1.add(Dense(10,activation='softmax'))
```

```python
x_train[0].shape
```

```python
model1.summary()
```

```python
(784*64)+64
```

```python
(64*64)+64
```

```python
lr=0.001
model1.compile(loss='categorical_crossentropy',
               optimizer=RMSprop(learning_rate=lr),
               metrics=['accuracy'])
```

```python
batch_size=128
epochs=30
```

```python
history=model1.fit(x_train,y_train,batch_size=batch_size,epochs=epochs,
                   verbose=1, validation_data=(x_test,y_test))
```

```python
model1.evaluate(x_test,y_test,verbose=1)
```

```python
score=model1.evaluate(x_test,y_test,verbose=0)
```

```python
print("Test loss:",score[0])
print("Test accuracy:",score[1])
```

```python
plt.figure(figsize=(7,9))
plt.plot(history.history['loss'],label="Train_loss")
plt.plot(history.history['val_loss'],label="Validation_loss")
plt.plot(history.history['accuracy'],label="Train_accuracy")
plt.plot(history.history['val_accuracy'],label="Validation_accuracy")
plt.yticks(np.arange(0.0,1.1,0.1))
plt.legend()
plt.show()
```
