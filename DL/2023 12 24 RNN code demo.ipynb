```python
docs = ['go india',
        'india india',
        'hip hip hurray',
        'vande mataram',
        'comeon india',
        'dhoni dhoni',
        'sachin sachin',
        'kohli kohli',
        'inquilab zindabad',
        'jeetega bhai jeetega india jeetega']
```

```python
from keras.preprocessing.text import Tokenizer
tokenizer=Tokenizer()
```

```python
tokenizer.fit_on_texts(docs)
```

```python
tokenizer.word_index
```

```python
sequences=tokenizer.texts_to_sequences(docs)
sequences
```

```python
from keras.utils import pad_sequences
sequences=pad_sequences(sequences,padding='post')
sequences
```

```python
from keras import Sequential
from keras.layers import Dense, SimpleRNN, Embedding, Flatten
```

```python
model1=Sequential()
model1.add(Embedding(15,output_dim=2,input_length=5))
model1.summary()
```

```python
model1.compile('adam','accuracy')
```

```python
model1.predict(sequences)
```

```python
from keras.datasets import imdb
```

```python
(X_train,y_train),(X_test,y_test)=imdb.load_data()
```

```python
X_train=pad_sequences(X_train,padding='post',maxlen=50)
X_test=pad_sequences(X_test,padding='post',maxlen=50)
```

```python
X_train.shape
```

```python
model2=Sequential()
model2.add(Embedding(10000,2))
model2.add(SimpleRNN(32,return_sequences=False))
model2.add(Dense(1,activation='sigmoid'))
```

```python
model2.summary()
```

```python
model2.compile(optimizer='adam',loss='binary_crossentropy',metrics=['acc'])
history=model2.fit(X_train,y_train,epochs=5,validation_data=(X_test,y_test))
```
