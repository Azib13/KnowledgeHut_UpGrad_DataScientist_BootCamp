```python
from prophet import Prophet
```

```python
op=Prophet()
```

```python
op.fit(df)
```

```python
forecast=op.make_future_dataframe(periods=700,freq="D")
```

```python
forecast
```

```python
forecast=op.predict(forecast)
forecast
```

```python
op.plot(forecast,xlabel='Date',ylabel='Potato price/q',figsize=(20,8),include_legend=True)
```

```python
from prophet.plot import plot_plotly, plot_components_plotly
plot_plotly(op, forecast)
```
