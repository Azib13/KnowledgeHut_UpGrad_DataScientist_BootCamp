```python
img=cv2.imread("/content/p2.jpg")
cv2_imshow(img)
```

```python
img=cv2.resize(img,(300,300))
cv2_imshow(img)
```

```python
# Edge Detection
sobel_x=cv2.Sobel(img,cv2.CV_8U,dx=1,dy=0,ksize=-1)
sobel_y=cv2.Sobel(img,cv2.CV_8U,dx=0,dy=1,ksize=-1)
sobel_f=cv2.bitwise_or(sobel_x,sobel_y)
la=cv2.Laplacian(img,cv2.CV_8U)
cv2_imshow(sobel_x)
cv2_imshow(sobel_y)
cv2_imshow(sobel_f)
cv2_imshow(la)
```

```python
a=np.ones([200,200,3])*255
a[:100,:]=[0,0,0]
cv2_imshow(a)
```

```python
sobel_y=cv2.Sobel(a,cv2.CV_8U,dx=0,dy=1,ksize=-1)
cv2_imshow(sobel_y)
```

```python
filter1=np.array([[-1,-1,-1],
                  [0,0,0],
                  [1,1,1]])
```

```python
op=cv2.filter2D(img,-1,filter1)
cv2_imshow(img)
cv2_imshow(op)
```

```python
op=cv2.filter2D(a,-1,filter1)
cv2_imshow(a)
cv2_imshow(op)
```
